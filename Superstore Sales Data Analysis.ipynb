# Exploratory Data Analysis

# 1. What is the Overall Sales Trend?

```python
superstore['order_date'].min()
```

```python
superstore['order_date'].max()
```

```python
#superstore('month_year')
superstore['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Creating a new column to fecth month and year from order date.
superstore['month_year'] = superstore['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
superstore['month_year'] 
```

```python
superstore.groupby('month_year').sum()
```

```python
# Grouping month year
superstore_Trend = superstore.groupby('month_year').sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(14,6))
plt.plot(superstore_Trend['month_year'], superstore_Trend['sales'], color='green')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

# 2. Which are the Top 10 products by sales? 

```python
superstore.head()
```

```python
superstore.groupby('product_name').sum()['sales']
```

```python
prod_sales = pd.DataFrame(superstore.groupby('product_name').sum()['sales'])
```

```python
# Sorting prod_sales column
prod_sales = prod_sales.sort_values('sales', ascending=False)
```

```python
# Top 10 products by sales
prod_sales[:10]
```

# 3. Which are the most selling products?

```python
# Grouping product name
superstore.groupby('product_name').sum()['quantity']
```

```python
prod_most_sell = pd.DataFrame(superstore.groupby('product_name').sum()['quantity'])
```

```python
prod_most_sell = prod_most_sell.sort_values('quantity', ascending = False)
```

```python
prod_most_sell[:10]
```
